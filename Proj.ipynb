```python
import os
import librosa
import numpy as np
import pandas as pd

# Define the emotion mapping
emotion_mapping = {
    1: 'boredom/sigh', 2: 'neutral/calm', 3: 'happy/laugh/gaggle', 4: 'sad/cry',
    5: 'angry/grunt/frustration', 6: 'fearful/scream/panic', 7: 'disgust/dislike/contempt',
    8: 'surprised/gasp/amazed', 9: 'excited', 10: 'pleasure', 11: 'pain/groan',
    12: 'disappointment/disapproval', 13: 'breath'
}

included_emotions = ['neutral/calm', 'happy/laugh/gaggle', 'surprised/gasp/amazed', 'excited', 'pleasure']

def map_emotions(emotion):
    return 1 if emotion in included_emotions else 0

# Function to extract MFCC features from audio file
def extract_mfcc(audio_path,max_length=1500):
    y, sr = librosa.load(audio_path, sr=None)  # Load audio file
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)  # Compute MFCC features
    mfccs_flat = np.hstack(mfccs)
    mfccs_processed = np.zeros(max_length)
    if len(mfccs_flat) >= max_length:
        mfccs_processed[:max_length] = mfccs_flat[:max_length]
    else:
        mfccs_processed[:len(mfccs_flat)] = mfccs_flat
    #if mfccs.shape[1] < max_length:
    #         pad_width = max_length - mfccs.shape[1]
    #         mfccs_processed = np.pad(mfccs, pad_width=((0, 0), (0, pad_width)))
    # else:
    #         mfccs_processed = mfccs[:, :max_length]

    feature=[]
    for i in range(len(mfccs_processed)):
        feature.append(np.mean(mfccs_processed[i]))
    return np.array(feature)

# Function to process audio files in a folder
def process_audio_folder(folder_path, max_iterations=None, min_iterations=None):
    data = []
    iteration_count = 0
    for filename in os.listdir(folder_path):
        if min_iterations is None or iteration_count>=min_iterations:
          if filename.endswith(".wav"):
              # Extract emotion from the filename based on the provided convention
              parts = filename.split('-')
              emotion_code = int(parts[2])  # Emotion code is the third part of the filename
              emotion = emotion_mapping.get(emotion_code, 'Unknown')  # Get emotion label
              label = map_emotions(emotion)  # Map emotion to binary label

              # Load the audio file and extract MFCC features
              audio_path = os.path.join(folder_path, filename)
              mfcc_features = extract_mfcc(audio_path)


              data.append((mfcc_features, label))

        iteration_count += 1
        if max_iterations is not None and iteration_count >= max_iterations:
                break

    # Convert data to DataFrame
    df = pd.DataFrame(data, columns=['mfcc_features', 'label'])
    return df

# Define the folder path
folder_path = ASVP_wav



```

```python
# Process audio files in the folder with a maximum of 100 iterations
audio_data = process_audio_folder(folder_path, max_iterations=1500)
validation_data=process_audio_folder(folder_path,max_iterations=1600,min_iterations=1501)
```

```python
from sklearn.model_selection import train_test_split
# X = np.vstack(audio_data['mfcc_features'].values)
X=np.vstack(audio_data['mfcc_features'])
y = audio_data['label'].to_numpy()

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=126)
```

```python
X_val=np.vstack(validation_data['mfcc_features'])
y_val = validation_data['label'].to_numpy()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(n_components=1)  # Adjust n_components as needed

# Fit and transform the data
X_train_lda = lda.fit_transform(X_train, y_train)
X_test_lda = lda.fit_transform(X_test, y_test)
```

```python
def norm(arr):
    # Transpose the array to work on columns
    arr = arr.T
    # Apply min-max normalization to each column
    normalized_cols = [(col - np.min(col)) / (np.max(col) - np.min(col)) for col in arr]
    # Transpose back to the original shape
    return np.array(normalized_cols).T
```
